# Container

```python
inputs = [1,8,6,2,5,4,8,3,7]
# inputs_trans = [(1,0),(8,1),(6,2),(2,3),(5,4),(4,5),(8,6),(3,7),(7,8)]
inputs_trans = []
```

```python
for i,j in enumerate(inputs):
    inputs_trans.append((i,j))
print(inputs_trans)    
```

```python
max_result = 0
for i in inputs_trans:
    w1, h1 = i
    # print("w1 : ", w1)
    # print("h1 : ", h1)
    for j in inputs_trans:
        w2, h2 = j
        # print("w2 : ", w2)
        # print("h2 : ", h2)
        result = min(h1,h2)*abs(w1-w2)
        if result > max_result:
            max_result = result
        
print(max_result)      

    
```

```python
print(inputs_trans)
```

```python
left = 0
right = len(inputs_trans) -1
max_area = 0
while left < right:
    width = right - left
    height = min(inputs[left],inputs[right])
    
    area = width*height
    if max_area < area:
        max_area = area
        
    if inputs[left] < inputs[right]:
        left += 1
    else:
        right -= 1
print(max_area)
    
    
```

# แบบฝึกหัด while loop

เป้าหมาย: ฝึกใช้ `while` เพื่อวนซ้ำเมื่อยังไม่จบเงื่อนไข เช่น นับเวลา รอคำตอบ ลองใหม่ สะสมคะแนน และจำลองสถานการณ์ในเกม

## กติกา

- แต่ละข้อให้เขียนโค้ดใน cell ที่เตรียมไว้
- ใช้ `while` อย่างน้อยข้อละ 1 ครั้ง
- ต้องมีการเปลี่ยนค่าตัวแปรใน loop เพื่อไม่ให้เกิด infinite loop
- ถ้าโจทย์ใช้ list จำลอง input ให้ใช้ index เดินไปทีละค่าแทนการใช้ `input()` เพื่อให้รันใน notebook ได้ง่าย
- ลองเปลี่ยนค่าตัวแปรทดสอบเองอย่างน้อย 2-3 กรณีต่อข้อ

## ข้อ 1: นับถอยหลังปลุกยานอวกาศ

ให้ใช้ `while` พิมพ์ตัวเลขถอยหลังจาก `countdown` ไปถึง 1 แล้วพิมพ์ `Ignition!`

ตัวอย่างถ้า `countdown = 5` ต้องได้

```text
5
4
3
2
1
Ignition!
```

````python
countdown = 5

# เขียนโค้ดตรงนี้
````

### เฉลยข้อ 1


````python
countdown = 5

while countdown > 0:
    print(countdown)
    countdown = countdown - 1
print("Ignition!")
````

## ข้อ 2: เติมพลังแบตเตอรี่หุ่นยนต์

หุ่นยนต์มีแบตเตอรี่เริ่มต้น `battery` และต้องชาร์จทีละ `charge_rate` จนถึงอย่างน้อย 100

ให้พิมพ์เปอร์เซ็นต์แบตเตอรี่ทุกครั้งหลังชาร์จ เช่น `Battery: 45%`

เมื่อเต็มแล้วให้พิมพ์ `Ready!`

````python
battery = 20
charge_rate = 25

# เขียนโค้ดตรงนี้
````

### เฉลยข้อ 2


````python
battery = 20
charge_rate = 25

while battery < 100:
    battery = battery + charge_rate
    if battery > 100:
        battery = 100
    print(f"Battery: {battery}%")
print("Ready!")
````

## ข้อ 3: ระบบล็อกอินลองรหัสผ่าน

มี list `attempts` จำลองรหัสที่ผู้ใช้ลองกรอก และรหัสจริงคือ `correct_password`

ให้ใช้ `while` ตรวจทีละรหัสจนกว่าจะเจอรหัสถูก หรือหมด attempt

- ถ้าถูกให้พิมพ์ `Login success`
- ถ้าผิดให้พิมพ์ `Wrong password`
- ถ้าลองหมดแล้วยังไม่ถูกให้พิมพ์ `Account locked`

````python
attempts = ["1111", "abcd", "gdd2026"]
correct_password = "gdd2026"

# เขียนโค้ดตรงนี้
````

### เฉลยข้อ 3


````python
attempts = ["1111", "abcd", "gdd2026"]
correct_password = "gdd2026"

index = 0
logged_in = False

while index < len(attempts) and not logged_in:
    if attempts[index] == correct_password:
        print("Login success")
        logged_in = True
    else:
        print("Wrong password")
    index = index + 1

if not logged_in:
    print("Account locked")
````

## ข้อ 4: เกมทายเลขแบบจำลอง

มีเลขคำตอบ `secret` และ list `guesses` จำลองเลขที่ผู้เล่นทาย

ให้ใช้ `while` ตรวจเลขที่ทายทีละตัว

- ถ้าทายน้อยไป พิมพ์ `Too low`
- ถ้าทายมากไป พิมพ์ `Too high`
- ถ้าถูก พิมพ์ `Correct!` แล้วหยุด loop
- ถ้าทายหมดแล้วยังไม่ถูก พิมพ์ `Out of guesses`

````python
secret = 18
guesses = [10, 25, 16, 18, 20]

# เขียนโค้ดตรงนี้
````

### เฉลยข้อ 4


````python
secret = 18
guesses = [10, 25, 16, 18, 20]

index = 0
found = False

while index < len(guesses) and not found:
    guess = guesses[index]
    if guess < secret:
        print("Too low")
    elif guess > secret:
        print("Too high")
    else:
        print("Correct!")
        found = True
    index = index + 1

if not found:
    print("Out of guesses")
````

## ข้อ 5: เก็บเหรียญจนซื้อไอเทมได้

ผู้เล่นเริ่มต้นมี `coins` เหรียญ และในแต่ละรอบได้เหรียญเพิ่มจาก list `rewards`

ให้ใช้ `while` เก็บเหรียญจนมีอย่างน้อย `price` หรือ rewards หมด

- ทุกครั้งที่ได้เหรียญ ให้พิมพ์ `Coins: ...`
- ถ้าซื้อได้ ให้พิมพ์ `Buy item`
- ถ้า rewards หมดแล้วยังซื้อไม่ได้ ให้พิมพ์ `Need more coins`

````python
coins = 15
price = 60
rewards = [10, 12, 8, 20]

# เขียนโค้ดตรงนี้
````

### เฉลยข้อ 5


````python
coins = 15
price = 60
rewards = [10, 12, 8, 20]

index = 0

while coins < price and index < len(rewards):
    coins = coins + rewards[index]
    print(f"Coins: {coins}")
    index = index + 1

if coins >= price:
    print("Buy item")
else:
    print("Need more coins")
````

## ข้อ 6: ลบพลังชีวิตมอนสเตอร์

มอนสเตอร์มีเลือด `monster_hp` และผู้เล่นโจมตีทีละ `damage`

ให้ใช้ `while` โจมตีจนเลือดมอนสเตอร์เหลือ 0 หรือน้อยกว่า

ทุกครั้งหลังโจมตีให้พิมพ์ `Monster HP: ...` โดยไม่ให้แสดงค่าติดลบ

เมื่อจบให้พิมพ์ `Monster defeated`

````python
monster_hp = 55
damage = 18

# เขียนโค้ดตรงนี้
````

### เฉลยข้อ 6


````python
monster_hp = 55
damage = 18

while monster_hp > 0:
    monster_hp = monster_hp - damage
    if monster_hp < 0:
        monster_hp = 0
    print(f"Monster HP: {monster_hp}")
print("Monster defeated")
````

## ข้อ 7: อ่านข้อความจนเจอคำหยุด

มี list `messages` จำลองข้อความที่เข้ามา ให้ใช้ `while` อ่านทีละข้อความจนเจอคำว่า `STOP`

- ถ้ายังไม่ใช่ `STOP` ให้พิมพ์ข้อความนั้น
- ถ้าเจอ `STOP` ให้หยุด loop แล้วพิมพ์ `End of messages`

````python
messages = ["hello", "keep going", "almost there", "STOP", "do not print"]

# เขียนโค้ดตรงนี้
````

### เฉลยข้อ 7


````python
messages = ["hello", "keep going", "almost there", "STOP", "do not print"]

index = 0

while index < len(messages) and messages[index] != "STOP":
    print(messages[index])
    index = index + 1

print("End of messages")
````

## ข้อ 8: สร้างเลข Fibonacci

ให้ใช้ `while` สร้างเลข Fibonacci จำนวน `limit` ตัวแรก

ตัวอย่างถ้า `limit = 7` ต้องได้ list `[0, 1, 1, 2, 3, 5, 8]`

สุดท้ายให้พิมพ์ list ที่ได้

````python
limit = 7

# เขียนโค้ดตรงนี้
````

### เฉลยข้อ 8


````python
limit = 7

fib = []
a = 0
b = 1

while len(fib) < limit:
    fib.append(a)
    next_number = a + b
    a = b
    b = next_number

print(fib)
````

## ข้อ 9: เครื่องถอนเงินแบบจำกัดธนบัตร

ต้องถอนเงินจำนวน `amount` โดยใช้ธนบัตร 100 บาทเท่านั้น

ให้ใช้ `while` นับจำนวนธนบัตรที่ต้องจ่าย

- ถ้า amount ไม่หารด้วย 100 ลงตัว ให้พิมพ์ `Invalid amount`
- ถ้าถอนได้ ให้พิมพ์จำนวนธนบัตร เช่น `Bills: 7`

````python
amount = 700

# เขียนโค้ดตรงนี้
````

### เฉลยข้อ 9


````python
amount = 700

if amount % 100 != 0:
    print("Invalid amount")
else:
    bills = 0
    while amount > 0:
        amount = amount - 100
        bills = bills + 1
    print(f"Bills: {bills}")
````

## ข้อ 10: Cooldown สกิลในเกม

สกิลมี cooldown เริ่มต้น `cooldown` วินาที ให้ใช้ `while` นับเวลาลดลงทีละ 1

- ระหว่าง cooldown ให้พิมพ์ `Skill ready in ...`
- เมื่อ cooldown เหลือ 0 ให้พิมพ์ `Skill ready!`

````python
cooldown = 4

# เขียนโค้ดตรงนี้
````

### เฉลยข้อ 10


````python
cooldown = 4

while cooldown > 0:
    print(f"Skill ready in {cooldown}")
    cooldown = cooldown - 1
print("Skill ready!")
````

## Challenge: จำลองร้านขายน้ำด้วย while loop

ให้สร้างระบบขายน้ำโดยมีตัวแปรเริ่มต้นอย่างน้อย 3 ตัว เช่น `stock`, `money`, `orders`

เงื่อนไขขั้นต่ำ

- ใช้ `while`
- มีการหยุด loop เมื่อของหมด หรือ order หมด
- มี `if/else` ตรวจว่าเงินพอซื้อหรือไม่
- ทุกครั้งที่ขายสำเร็จต้องลด stock และเพิ่มยอดขายรวม
- สุดท้ายพิมพ์ยอดขายรวมและจำนวนสินค้าที่เหลือ

````python
stock = 5
price = 20
orders = [20, 10, 50, 20, 100, 20]

# เขียนโจทย์และโค้ดของตัวเองตรงนี้
````

### ตัวอย่างเฉลย Challenge


````python
stock = 5
price = 20
orders = [20, 10, 50, 20, 100, 20]
index = 0
sales = 0

while stock > 0 and index < len(orders):
    money = orders[index]
    if money >= price:
        print("Sold 1 drink")
        stock = stock - 1
        sales = sales + price
    else:
        print("Not enough money")
    index = index + 1

print(f"Total sales: {sales}")
print(f"Stock left: {stock}")
````

## ข้อ 1: นับถอยหลังปล่อยจรวด

ให้ใช้ `for` พิมพ์ตัวเลขถอยหลังจาก `start` ไปถึง 1 แล้วพิมพ์ `Launch!`

ตัวอย่าง: ถ้า `start = 5` ต้องได้

```text
5
4
3
2
1
Launch!
```

````python
start = 5

# เขียนโค้ดตรงนี้
````

### เฉลยข้อ 1


````python
start = 5

for number in range(start, 0, -1):
    print(number)
print("Launch!")
````

## ข้อ 3: รวมพลังเหรียญในเกม

มี list `coins` เก็บจำนวนเหรียญที่เก็บได้แต่ละด่าน ให้ใช้ `for` รวมเหรียญทั้งหมด แล้วพิมพ์ `Total coins: ...`

## ข้อ 4: ตรวจคะแนนที่ผ่านเกณฑ์

มี list `scores` ให้พิมพ์เฉพาะคะแนนที่มากกว่าหรือเท่ากับ 50

จากนั้นพิมพ์จำนวนคนที่ผ่าน เช่น `Passed: 3 people`

````python
scores = [72, 48, 91, 33, 50, 64]

# เขียนโค้ดตรงนี้
````

### เฉลยข้อ 4


````python
scores = [72, 48, 91, 33, 50, 64]

passed_count = 0
for score in scores:
    if score >= 50:
        print(score)
        passed_count = passed_count + 1
print(f"Passed: {passed_count} people")
````

## ข้อ 5: เครื่องกรองคำต้องห้าม

มี list `words` ให้สร้าง list ใหม่ชื่อ `clean_words` ที่เก็บเฉพาะคำที่ไม่ใช่ `"spam"`

สุดท้ายพิมพ์ `clean_words`

````python
words = ["cat", "spam", "game", "spam", "python", "art"]

# เขียนโค้ดตรงนี้
````

### เฉลยข้อ 5


````python
words = ["cat", "spam", "game", "spam", "python", "art"]

clean_words = []
for word in words:
    if word != "spam":
        clean_words.append(word)
print(clean_words)
````

## ข้อ 6: นับสระในประโยค

ให้ใช้ `for` นับจำนวนตัวอักษรภาษาอังกฤษที่เป็นสระในตัวแปร `sentence`

สระคือ `a`, `e`, `i`, `o`, `u` และให้นับทั้งตัวพิมพ์เล็ก/ใหญ่

พิมพ์ผลลัพธ์เช่น `Vowels: 5`

````python
sentence = "Game Design is fun"

# เขียนโค้ดตรงนี้
````

### เฉลยข้อ 6


````python
sentence = "Game Design is fun"

vowels = "aeiou"
count = 0
for char in sentence.lower():
    if char in vowels:
        count = count + 1
print(f"Vowels: {count}")
````

## ข้อ 8: แปลง inventory เป็นข้อความ

มี dictionary `inventory` เก็บชื่อไอเทมและจำนวน ให้ใช้ `for` พิมพ์ทุกไอเทมในรูปแบบ `item: amount`

เช่น `potion: 3`

````python
inventory = {"potion": 3, "sword": 1, "coin": 120, "map": 2}

# เขียนโค้ดตรงนี้
````

### เฉลยข้อ 8

## ข้อ 9: แผนที่คะแนนในตาราง 2 มิติ

มี nested list `score_map` ให้ใช้ nested `for` รวมคะแนนทั้งหมดในตาราง แล้วพิมพ์ `Total score: ...`

คำใบ้: ใช้ for ซ้อน for เพื่อเข้าไปดูตัวเลขแต่ละช่อง

````python
score_map = [
    [5, 10, 15],
    [20, 0, 5],
    [7, 8, 9]
]

# เขียนโค้ดตรงนี้
````

### เฉลยข้อ 9


````python
score_map = [
    [5, 10, 15],
    [20, 0, 5],
    [7, 8, 9]
]

total = 0
for row in score_map:
    for score in row:
        total = total + score
print(f"Total score: {total}")
````

## ข้อ 10: สร้างแถบพลังชีวิต

จากตัวแปร `hp` และ `max_hp` ให้สร้าง string แถบพลัง 10 ช่อง

- ช่องที่มีพลังให้ใช้ `#`
- ช่องที่หายไปให้ใช้ `-`

ตัวอย่าง ถ้า hp เป็นครึ่งหนึ่งของ max_hp ให้ได้ `#####-----`

ให้พิมพ์ในรูปแบบ `HP: #####-----`

````python
hp = 35
max_hp = 50

# เขียนโค้ดตรงนี้
````

### เฉลยข้อ 10


````python
hp = 35
max_hp = 50

filled_slots = int((hp / max_hp) * 10)
bar = ""
for i in range(10):
    if i < filled_slots:
        bar = bar + "#"
    else:
        bar = bar + "-"
print(f"HP: {bar}")
````

## Challenge: สร้างลาย pixel art ด้วย for loop

ให้สร้างลายจากตัวแปร `size` โดยใช้ nested `for` และพิมพ์ออกมาเป็นรูปสี่เหลี่ยม

เงื่อนไขขั้นต่ำ

- ใช้ nested `for`
- มี `if/else` เพื่อเลือกว่าจะพิมพ์สัญลักษณ์อะไร
- ใช้ string สะสมแต่ละแถวก่อน `print`
- ลายต้องเปลี่ยนขนาดได้เมื่อเปลี่ยนค่า `size`

ตัวอย่างลายที่ทำได้: กรอบสี่เหลี่ยม, เส้นทแยงมุม, ลาย X, ลายกระดานหมากรุก

````python
size = 7

# เขียนโจทย์และโค้ดของตัวเองตรงนี้
````

### ตัวอย่างเฉลย Challenge


````python
# ตัวอย่าง: สร้างลาย X ขนาด size x size
size = 7

for row in range(size):
    line = ""
    for col in range(size):
        if row == col or row + col == size - 1:
            line = line + "#"
        else:
            line = line + "."
    print(line)
````

## ข้อ 1: ประตูร้านราเมงอัจฉริยะ

ร้านราเมงมีระบบให้คิวอัตโนมัติจากจำนวนลูกค้าและสถานะสมาชิก

- ถ้า `customers` มากกว่า 8 คน และ `is_member` เป็น `True` ให้พิมพ์ `รับบัตรคิวด่วน`
- ถ้า `customers` มากกว่า 8 คน แต่ไม่ใช่สมาชิก ให้พิมพ์ `รอคิวปกติ`
- ถ้า `customers` อยู่ระหว่าง 1 ถึง 8 ให้พิมพ์ `เข้าร้านได้`
- ถ้า `customers` เท่ากับ 0 ให้พิมพ์ `ร้านว่าง`

```python
customers = 12
is_member = True

if customers > 8:
    if is_member:
        print("รับบัตรคิวด่วน")
    else:
        print("รอคิวปกติ")
elif 1 <= customers <= 8:
    print("เข้าร้านได้")
else:
    print("ร้านว่าง")
```

## ข้อ 2: เครื่องคัดแยกพัสดุลึกลับ

ให้เขียนโปรแกรมคัดแยกพัสดุจาก `weight` และ `fragile`

- ถ้าน้ำหนักมากกว่า 20 kg ให้พิมพ์ `ส่งด้วยรถบรรทุก`
- ถ้าน้ำหนัก 5 ถึง 20 kg และ fragile เป็น `True` ให้พิมพ์ `ส่งแบบกันกระแทก`
- ถ้าน้ำหนัก 5 ถึง 20 kg และ fragile เป็น `False` ให้พิมพ์ `ส่งแบบมาตรฐาน`
- ถ้าน้ำหนักน้อยกว่า 5 kg ให้พิมพ์ `ส่งแบบซองหรือกล่องเล็ก`

```python
weight = 7
fragile = True

if weight > 20:
    print("ส่งด้วยรถบรรทุก")
elif 5 <= weight <=20 :
    if fragile :
        print("ส่งแบบกันกระอทก")
    elif fragile:
        print("ส่งแบบมาตรฐาน")
else:
    print("ส่งแบบซองหรือกล่องเล็ก")
```

## ข้อ 3: ระบบเลือกอาวุธในเกม RPG

ตัวละครมี `level`, `mana`, และ `enemy_type`

- ถ้าเจอ `dragon` และ level ตั้งแต่ 10 ขึ้นไป และ mana ตั้งแต่ 50 ขึ้นไป ให้พิมพ์ `ใช้เวทน้ำแข็ง`
- ถ้าเจอ `dragon` แต่เงื่อนไขไม่พอ ให้พิมพ์ `หนีไปตั้งหลัก`
- ถ้าเจอ `slime` ให้พิมพ์ `ใช้ดาบไม้`

```python
enemy_type = "dragon"
mana = 90
level = 11
if enemy_type == "dragon":
    if level >= 10 and mana >= 50:
        print("ใช้เวทน้ำแข็ง")
    else:
        print("หนีไปตั้งหลัก")
elif enemy_type == "slime":
    print("ใช้ดาบไม้")
else:
    print("ประเมินสถานการณ์ก่อนโจมตี")
#hint for loop and enumerate
```

## ข้อ 4: ตู้กดเครื่องดื่มตามอารมณ์

จากตัวแปร `mood` และ `temperature`

- ถ้า mood เป็น `sleepy` ให้แนะนำ `กาแฟเย็น`
- ถ้า mood เป็น `happy` และ temperature มากกว่า 30 ให้แนะนำ `น้ำมะนาวโซดา`
- ถ้า mood เป็น `happy` และ temperature ไม่เกิน 30 ให้แนะนำ `ชาเขียว`
- ถ้า mood เป็น `stressed` ให้แนะนำ `โกโก้ร้อน`
- ถ้าไม่ตรงกรณีใด ให้แนะนำ `น้ำเปล่า`

```python
alphabet = "abcdef"
result = []
for i in alphabet:
    for j in alphabet:
        result.append(i + j)
print(result)
    
```

## ข้อ 5: ห้องสมุดคืนหนังสือ

ห้องสมุดคิดค่าปรับจากจำนวนวันที่คืนช้า `late_days`

- 0 วัน: พิมพ์ `ไม่มีค่าปรับ`
- 1 ถึง 3 วัน: ค่าปรับวันละ 5 บาท
- 4 ถึง 7 วัน: ค่าปรับวันละ 10 บาท
- มากกว่า 7 วัน: ค่าปรับวันละ 20 บาท และพิมพ์เพิ่มว่า `กรุณาติดต่อบรรณารักษ์`

ให้พิมพ์ค่าปรับรวมออกมา เช่น `ค่าปรับ 30 บาท`

```python
late_days = 6

if late_days == 0 :
    print("ไม่มีค่าปรับ")
elif late_days <= 3:
    fine = late_days*5
    print(f"ค่าปรับ {fine} บาท")
elif 4<= late_days <= 7:
    fine = late_days*10
    print(f"ค่าปรับ {fine} บาท")
else:
    fine = late_days*20
    print(f"ค่าปรับ {fine} บาท")
```

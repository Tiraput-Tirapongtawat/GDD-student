- ถ้า `customers` มากกว่า 8 คน แต่ไม่ใช่สมาชิก ให้พิมพ์ `รอคิวปกติ`

- ถ้าน้ำหนักมากกว่า 20 kg ให้พิมพ์ `ส่งด้วยรถบรรทุก`
- ถ้าน้ำหนัก 5 ถึง 20 kg และ fragile เป็น `True` ให้พิมพ์ `ส่งแบบกันกระแทก`
- ถ้าน้ำหนัก 5 ถึง 20 kg และ fragile เป็น `False` ให้พิมพ์ `ส่งแบบมาตรฐาน`
- ถ้าน้ำหนักน้อยกว่า 5 kg ให้พิมพ์ `ส่งแบบซองหรือกล่องเล็ก`

### เฉลยข้อ 2


```python
weight = 7.5
fragile = True

if weight > 20:
    print("ส่งด้วยรถบรรทุก")
elif weight >= 5:
    if fragile:
        print("ส่งแบบกันกระแทก")
    else:
        print("ส่งแบบมาตรฐาน")
else:
    print("ส่งแบบซองหรือกล่องเล็ก")
```

## ข้อ 4: ตู้กดเครื่องดื่มตามอารมณ์

จากตัวแปร `mood` และ `temperature`

- ถ้า mood เป็น `sleepy` ให้แนะนำ `กาแฟเย็น`
- ถ้า mood เป็น `happy` และ temperature มากกว่า 30 ให้แนะนำ `น้ำมะนาวโซดา`

- ถ้า mood เป็น `stressed` ให้แนะนำ `โกโก้ร้อน`

## ข้อ 5: ห้องสมุดคืนหนังสือ

ห้องสมุดคิดค่าปรับจากจำนวนวันที่คืนช้า `late_days`

- 0 วัน: พิมพ์ `ไม่มีค่าปรับ`
- 1 ถึง 3 วัน: ค่าปรับวันละ 5 บาท
- 4 ถึง 7 วัน: ค่าปรับวันละ 10 บาท
- มากกว่า 7 วัน: ค่าปรับวันละ 20 บาท และพิมพ์เพิ่มว่า `กรุณาติดต่อบรรณารักษ์`

ให้พิมพ์ค่าปรับรวมออกมา เช่น `ค่าปรับ 30 บาท`

```python
late_days = 6

# เขียนโค้ดตรงนี้
```

### เฉลยข้อ 5


```python
late_days = 6

if late_days == 0:
    print("ไม่มีค่าปรับ")
elif late_days <= 3:
    fine = late_days * 5
    print(f"ค่าปรับ {fine} บาท")
elif late_days <= 7:
    fine = late_days * 10
    print(f"ค่าปรับ {fine} บาท")
else:
    fine = late_days * 20
    print(f"ค่าปรับ {fine} บาท")
    print("กรุณาติดต่อบรรณารักษ์")
```

## ข้อ 6: ระบบตรวจรหัสผ่านแบบนักสืบ

ให้ตรวจ `password` ตามเงื่อนไขนี้

- ถ้าความยาวน้อยกว่า 8 ตัวอักษร ให้พิมพ์ `สั้นเกินไป`
- ถ้าไม่มีตัวเลขเลย ให้พิมพ์ `ต้องมีตัวเลขอย่างน้อย 1 ตัว`
- ถ้ามีคำว่า `password` อยู่ในรหัส ให้พิมพ์ `เดาง่ายเกินไป`
- ถ้าผ่านทุกเงื่อนไข ให้พิมพ์ `รหัสผ่านใช้ได้`

คำใบ้: ใช้ loop ตรวจว่ามีตัวเลขหรือไม่ หรือใช้ `.isdigit()` กับตัวอักษรแต่ละตัว

```python
password = "gdd2026"

# เขียนโค้ดตรงนี้
```

### เฉลยข้อ 6


```python
password = "gdd2026"

has_number = False
for char in password:
    if char.isdigit():
        has_number = True

if len(password) < 8:
    print("สั้นเกินไป")
elif not has_number:
    print("ต้องมีตัวเลขอย่างน้อย 1 ตัว")
elif "password" in password.lower():
    print("เดาง่ายเกินไป")
else:
    print("รหัสผ่านใช้ได้")
```

## ข้อ 7: เครื่องทำนายสภาพอากาศจากข้อมูลเซนเซอร์

ใช้ `humidity`, `cloud_level`, และ `wind_speed`

- ถ้า humidity มากกว่า 80 และ cloud_level มากกว่า 7 ให้พิมพ์ `ฝนอาจตกหนัก`
- ถ้า humidity มากกว่า 60 และ cloud_level มากกว่า 4 ให้พิมพ์ `มีโอกาสฝนตก`
- ถ้า wind_speed มากกว่า 40 ให้พิมพ์ `ลมแรง ระวังสิ่งของปลิว`
- ถ้าไม่เข้าเงื่อนไขข้างบน ให้พิมพ์ `อากาศปกติ`

ลองคิดด้วยว่า ถ้า `wind_speed` แรงมากแต่ฝนก็ตกด้วย ควรให้โปรแกรมแสดงข้อความแบบไหน

```python
humidity = 85
cloud_level = 8
wind_speed = 45

# เขียนโค้ดตรงนี้
```

### เฉลยข้อ 7


```python
humidity = 85
cloud_level = 8
wind_speed = 45

if humidity > 80 and cloud_level > 7:
    print("ฝนอาจตกหนัก")
    if wind_speed > 40:
        print("ลมแรง ระวังสิ่งของปลิว")
elif humidity > 60 and cloud_level > 4:
    print("มีโอกาสฝนตก")
    if wind_speed > 40:
        print("ลมแรง ระวังสิ่งของปลิว")
elif wind_speed > 40:
    print("ลมแรง ระวังสิ่งของปลิว")
else:
    print("อากาศปกติ")
```

## ข้อ 9: ระบบให้เกรดแบบมีเงื่อนไขพิเศษ

ใช้ `score` และ `submitted_project`

- score 80 ขึ้นไป: เกรด A
- score 70 ถึง 79: เกรด B
- score 60 ถึง 69: เกรด C
- score 50 ถึง 59: เกรด D
- น้อยกว่า 50: เกรด F
- แต่ถ้า `submitted_project` เป็น `False` และเกรดเดิมไม่ใช่ F ให้ลดเกรดลง 1 ขั้น เช่น A เป็น B, B เป็น C

ให้พิมพ์เกรดสุดท้าย เช่น `Final grade: C`

```python
score = 82
submitted_project = False

# เขียนโค้ดตรงนี้
```

### เฉลยข้อ 9


```python
score = 82
submitted_project = False

if score >= 80:
    grade = "A"
elif score >= 70:
    grade = "B"
elif score >= 60:
    grade = "C"
elif score >= 50:
    grade = "D"
else:
    grade = "F"

if not submitted_project and grade != "F":
    if grade == "A":
        grade = "B"
    elif grade == "B":
        grade = "C"
    elif grade == "C":
        grade = "D"
    else:
        grade = "F"

print(f"Final grade: {grade}")
```

## ข้อ 10: ระบบปลดล็อกห้องทดลองลับ

```python
role = "student"
has_keycard = True
code = "GDD-42"
emergency_mode = False

# เขียนโค้ดตรงนี้
```

### เฉลยข้อ 10


```python
role = "student"
has_keycard = True
code = "GDD-42"
emergency_mode = False

if emergency_mode:
    if role == "admin":
        print("เปิดประตูฉุกเฉิน")
    else:
        print("ไม่มีสิทธิ์เปิดโหมดฉุกเฉิน")
else:
    if has_keycard and code == "GDD-42":
        print("เปิดประตู")
    elif has_keycard:
        print("รหัสผิด")
    else:
        print("ต้องใช้บัตรผ่าน")
```

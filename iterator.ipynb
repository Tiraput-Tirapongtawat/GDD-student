## Iterator Python

ตัววนซ้ำมีอยู่แทบทุกที่ใน Python มันถูกใช้อย่างแนบเนียนใน `for` loop, comprehension, generator และอื่น ๆ แม้เราจะไม่ค่อยสังเกตก็ตาม

ตัววนซ้ำใน Python ก็คือ **[object](https://github.com/milaan9/06_Python_Object_Class/blob/main/002_Python_Classes_and_Objects.ipynb)** ที่สามารถถูกวนซ้ำได้ เป็น object ที่จะคืนข้อมูลกลับมาทีละหนึ่งสมาชิก

ในเชิงเทคนิค **อ็อบเจ็กต์ตัววนซ้ำ** ใน Python ต้อง implement เมธอดพิเศษ 2 ตัว คือ **`__iter__()`** และ **`__next__()`** ซึ่งรวมกันเรียกว่า **โปรโตคอลของตัววนซ้ำ**

object จะถูกเรียกว่า **iterable** ถ้าเราสามารถดึงตัววนซ้ำออกมาจากมันได้ container แบบ built-in ส่วนใหญ่ใน Python เช่น **[string](https://github.com/milaan9/02_Python_Datatypes/blob/main/002_Python_String.ipynb)**, **[list](https://github.com/milaan9/02_Python_Datatypes/blob/main/003_Python_List.ipynb)**, **[tuple](https://github.com/milaan9/02_Python_Datatypes/blob/main/004_Python_Tuple.ipynb)** เป็นต้น ล้วนเป็น iterable

ฟังก์ชัน **`iter()`** (ซึ่งภายในจะเรียกเมธอด **`__iter__()`**) จะคืนค่าตัววนซ้ำจาก object เหล่านี้

```python
number_list = [5,4,3,2,4,5,68,9,9,10]

my_iter = iter(number_list)
```

```python
print(next(my_iter))
print(my_iter.__next__())
```

```python
class PowTwo:
    def __init__(self, max):
        self.max = max
        print("Test class")
    # def test(self):
    #     print("test function in class")
    def __iter__(self):
        self.n = 0
        return self
    def __next__(self):
        
        print(self.n)
        if self.n < self.max:
            self.n = self.n ** 2
            self.n += 1
        else:
            print("Finished")
            raise StopIteration
            
```

```python
c = PowTwo(max = 5)
```

```python
c.__iter__()
```

```python
c.__next__()
```

4. ลองทดสอบด้วย `Countdown(5)` และคาดหวังผลลัพธ์เป็น `5, 4, 3, 2, 1, 0`

```python
class Countdown:
    
    def __init__(self,start):
        self.start = start
    def __iter__(self):
        return self
    def __next__(self):
        self.start -= 1
        print(self.start)
        
        if self.start == 0 :
            print("Finish")
            raise StopIteration
```

```python
a = Countdown(5)
```

```python
a.__iter__()
```

```python
a.__next__()
```

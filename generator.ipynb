```python
def gen_str(word):
    length = len(word)
    for i in range(length-1,-1,-1):
        yield word[i] + "-"
        
```

```python
b = gen_str("gdd_coding")
```

```python
for i in gen_str("abcdefghijklmnop"):
    print(i)
# next(b)
```

```python
def PowTwoGen(max = 0):
    n = 0 
    while n < max:
        yield n**2
        n += 1
```

```python
b = PowTwoGen(max = 50)
```

```python
for j in PowTwoGen(max = 70):
    print(j)
```

```python
items =["Cat", "Dog","Dragon","Robot","King","Superman"]
```

```python
import random
def gacha_machine (items):
    items_copy = items[:]
    random.shuffle(items_copy)
    # print(items_copy)
    for i in items_copy:
        
        yield i
        
    yield "This gacha machine is empty."
    
```

```python
a = gacha_machine(items)
```

```python
for i in a :
    print(i)
```

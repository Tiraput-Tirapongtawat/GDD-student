# RegEx

**Reg**ular **Ex**pression (RegEx) คือชุดอักขระที่กำหนดรูปแบบสำหรับค้นหา เช่น

```python
>>> ^a...s$
```

โค้ดด้านบนกำหนดรูปแบบ RegEx โดยรูปแบบนี้หมายถึง: **สตริงความยาวห้าตัวอักษรที่ขึ้นต้นด้วย `a` และลงท้ายด้วย `s`**

รูปแบบที่กำหนดด้วย RegEx สามารถนำไปจับคู่กับสตริงได้

| นิพจน์ | สตริง | ตรงกันหรือไม่ | 
|:----| :--- |:--- |
|               |  **`abs`** | ไม่ตรงกัน | 
|               |  **`alias`** | ตรงกัน |
| **`^a...s$`** |  **`abyss`** | ตรงกัน |
|               |  **`Alias`** | ไม่ตรงกัน |
|               |  **`An abacus`** | ไม่ตรงกัน |

## อักขระพิเศษ (MetaCharacters)

Metacharacters คืออักขระที่ RegEx engine ตีความแบบพิเศษ รายการ metacharacters มีดังนี้:

###  **`[] . ^ $ * + ? {} () \\ |`**

ตัวอย่างการใช้งานครบทั้ง 11 ตัว:

| ลำดับ | Metacharacter | ความหมาย | ตัวอย่าง Pattern | ข้อความตัวอย่าง | ผลที่จับคู่ได้ |
|:---:|:---|:---|:---|:---|:---|
| 1 | **`[]`** | กำหนดชุดอักขระที่ยอมรับ | **`[Pp]ython`** | **`Python and python`** | **`Python`**, **`python`** |
| 2 | **`.`** | แทนอักขระใดก็ได้ 1 ตัว ยกเว้นขึ้นบรรทัดใหม่ | **`c.t`** | **`cat cut c9t`** | **`cat`**, **`cut`**, **`c9t`** |
| 3 | **`^`** | จับคู่ที่จุดเริ่มต้นของสตริง | **`^Hello`** | **`Hello world`** | **`Hello`** |
| 4 | **`$`** | จับคู่ที่จุดสิ้นสุดของสตริง | **`end$`** | **`the end`** | **`end`** |
| 5 | **`*`** | ซ้ำ 0 ครั้งหรือมากกว่า | **`ab*c`** | **`ac abc abbc`** | **`ac`**, **`abc`**, **`abbc`** |
| 6 | **`+`** | ซ้ำ 1 ครั้งหรือมากกว่า | **`ab+c`** | **`ac abc abbc`** | **`abc`**, **`abbc`** |
| 7 | **`?`** | มีหรือไม่มีก็ได้ 1 ครั้ง | **`colou?r`** | **`color colour`** | **`color`**, **`colour`** |
| 8 | **`{}`** | กำหนดจำนวนครั้งที่ซ้ำ | **`\d{3}`** | **`ID 123 and 4567`** | **`123`**, **`456`** |
| 9 | **`()`** | จัดกลุ่ม pattern | **`(ha)+`** | **`ha haha hahaha`** | **`ha`**, **`haha`**, **`hahaha`** |
| 10 | **`\`** | escape อักขระพิเศษ หรือใช้กับ special sequence | **`\.`** | **`version 3.10`** | **`.`** |
| 11 | **`|`** | เลือกอย่างใดอย่างหนึ่ง (or) | **`cat\|dog`** | **`cat and dog`** | **`cat`**, **`dog`** |

````python
import re

pattern = "^a...s$"

test_string = "abmds"

result = re.match(pattern, test_string)

print(result)

if result:
    print( "Found this pattern")

````

### เมธอด RegEx ใน Python

ในการค้นหารูปแบบ เราใช้ชุดอักขระและเมธอดต่าง ๆ ของ **`re`** ที่ช่วยค้นหาการจับคู่ในสตริง

* **`re.findall`**: คืนค่า list ที่มีการจับคู่ทั้งหมด
* **`re.split`**: รับสตริง แยกสตริงที่ตำแหน่งที่จับคู่ได้ แล้วคืนค่า list
* **`re.sub`**: แทนที่การจับคู่หนึ่งครั้งหรือหลายครั้งภายในสตริง
* **`re.search`**: คืนค่า match object หากพบการจับคู่ที่ตำแหน่งใดก็ได้ในสตริง รวมถึงสตริงหลายบรรทัด
* **`re.match()`**: ค้นหาเฉพาะตอนต้นของบรรทัดแรกของสตริง และคืนค่า matched object หากพบ ไม่เช่นนั้นคืนค่า **`None`**

ในการใช้งาน เราต้อง import โมดูลก่อน

````python
import re

string = "hello sdkgjdksgjn 345 lkflsndfjkn 5969"

pattern = '\d{3}'

result = re.findall(pattern, string)

print(result)
````

````python
re.split(pattern, string)
````

## Lists

ใน Python มีชนิดข้อมูลแบบ collection หลัก ๆ อยู่ 4 แบบ:

- **List**: เก็บข้อมูลแบบมีลำดับ แก้ไขได้ และมีค่าซ้ำได้
- **Tuple**: เก็บข้อมูลแบบมีลำดับ แต่แก้ไขไม่ได้ และมีค่าซ้ำได้
- **Set**: เก็บข้อมูลแบบไม่มีลำดับ ไม่มี index และไม่มีค่าซ้ำ แต่เพิ่มสมาชิกใหม่ได้
- **Dictionary**: เก็บข้อมูลเป็นคู่ key-value แก้ไขได้ และ key ต้องไม่ซ้ำกัน

List คือ collection ที่เก็บข้อมูลหลายค่าไว้ในตัวแปรเดียว ข้อมูลใน list จะมีลำดับ และสามารถแก้ไข เพิ่ม ลบ หรือเรียงลำดับได้ List อาจว่างเปล่า หรือเก็บข้อมูลหลายชนิดปนกันก็ได้

## การเข้าถึงสมาชิกด้วย Positive Indexing

index ของ list เริ่มจาก `0` สมาชิกตัวแรกคือ index `0` ตัวถัดไปคือ index `1` และเพิ่มไปเรื่อย ๆ

![List index](/Users/tiraputtirapongtawat/Documents/GDD/GDD-student/image/list_index.png)


```python
print(fruits[2])
print(fruits[3])

last_index = len(fruits) -1
print(fruits[last_index])
```

## การเข้าถึงสมาชิกด้วย Negative Indexing

negative index คือการนับจากท้าย list โดย `-1` หมายถึงสมาชิกตัวสุดท้าย, `-2` คือตัวรองสุดท้าย

![List negative indexing](/Users/tiraputtirapongtawat/Documents/GDD/GDD-student/image/list_negative_indexing.png)


```python
print(fruits[-1])
```

```python
fruits = ["banana", "orange", "mango", "lemon"] # start, stop,step

print(fruits[0:4])
print(fruits[0:4:2])
print(fruits[1:4:1])
print(fruits[::]) 

```

```python
fruits = ["banana", "orange", "mango", "lemon"]  # start, stop,step

print(fruits[-3])
print(fruits[-3:-1])
print(fruits[-3:])
print(fruits[::-1])
```

```python
monsters = ["slime", "bat", "ghost", "dragon", "zombie"]


while len(monsters) > 0:
    target = input("จะโจมตีมอนสเตอร์ตัวไหน: ")
    if target in monsters:
        print("คุณสามารถโจมตี moster ได้")
        monsters.remove(target)
        print(len(monsters))
    else :
        print("ไม่พบ monster ที่เป็น target")
else:
    print("Task completed")
    
```

```python
pizza_toppings = [
    "ชีส",
    "เปปเปอโรนี",
    "แฮม",
    "เบคอน",
    "เห็ด",
    "สับปะรด",
    "ไส้กรอก",
    "ซีฟู้ด",
    "ผักโขม",
    "มะกอก",
    "มะเขือเทศ",
    "หอมใหญ่",
]

#โจทย์ สั่งหน้าพิซซ่า ให้ลูกค้าสั่งจนกว่าลูกค้าจะ done 

# repeat order ที่ลูกค้าสั่ง
orders = []
while True:
    
    order_from_customer =  input("หน้าพิซซ่าที่ต้องการสั่ง : ")
    if order_from_customer in pizza_toppings:
        print("สั่งพิซซ่า สำเร็จ")
        orders.append(order_from_customer)
    else:
        print("สั่งพิซซ่า ไม่สำเร็จ")
        
    
    if order_from_customer == "done":
        break
    
    
print(orders)
    
    
```

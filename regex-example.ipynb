# Python RegEx

ในบทเรียนนี้ คุณจะได้เรียนรู้เกี่ยวกับ regular expressions (RegEx) และใช้โมดูล `re` ของ Python เพื่อทำงานกับ RegEx ผ่านตัวอย่างประกอบ

**Reg**ular **Ex**pression (RegEx) คือชุดอักขระที่กำหนดรูปแบบสำหรับค้นหา เช่น

```python
>>> ^a...s$
```

โค้ดด้านบนกำหนดรูปแบบ RegEx โดยรูปแบบนี้หมายถึง: **สตริงความยาวห้าตัวอักษรที่ขึ้นต้นด้วย `a` และลงท้ายด้วย `s`**

รูปแบบที่กำหนดด้วย RegEx สามารถนำไปจับคู่กับสตริงได้

| นิพจน์ | สตริง | ตรงกันหรือไม่ | 
|:----| :--- |:--- |
|               |  **`abs`** | ไม่ตรงกัน | 
|               |  **`alias`** | ตรงกัน |
| **`^a...s$`** |  **`abyss`** | ตรงกัน |
|               |  **`Alias`** | ไม่ตรงกัน |
|               |  **`An abacus`** | ไม่ตรงกัน |

Python มีโมดูลชื่อ **`re`** สำหรับทำงานกับ RegEx ตัวอย่างเช่น:

````python
import re

pattern = '^a...s$' #จะมีตัวตรงกลางได้ 3 ตัว
test_string = 'afyss'
result = re.match(pattern, test_string)

if result:
    print("ค้นหาสำเร็จ")
else:
    print("ค้นหาไม่สำเร็จ")	
````

ในที่นี้ เราใช้ฟังก์ชัน **`re.match()`** เพื่อค้นหา **`pattern`** ภายใน **`test_string`** เมธอดนี้จะคืนค่า match object หากค้นหาสำเร็จ หากไม่สำเร็จจะคืนค่า **`None`**

ยังมีฟังก์ชันอื่น ๆ อีกหลายตัวในโมดูล **`re`** สำหรับทำงานกับ RegEx ก่อนจะไปดูฟังก์ชันเหล่านั้น เรามาทำความเข้าใจ regular expressions กันก่อน

## การกำหนด Pattern ด้วย RegEx

การกำหนด regular expressions จะใช้ metacharacters ในตัวอย่างด้านบน **`^`** และ **`$`** คือ metacharacters

## อักขระพิเศษ (MetaCharacters)

Metacharacters คืออักขระที่ RegEx engine ตีความแบบพิเศษ รายการ metacharacters มีดังนี้:

###  **`[] . ^ $ * + ? {} () \\ |`**

ตัวอย่างการใช้งานครบทั้ง 11 ตัว:

| ลำดับ | Metacharacter | ความหมาย | ตัวอย่าง Pattern | ข้อความตัวอย่าง | ผลที่จับคู่ได้ |
|:---:|:---|:---|:---|:---|:---|
| 1 | **`[]`** | กำหนดชุดอักขระที่ยอมรับ | **`[Pp]ython`** | **`Python and python`** | **`Python`**, **`python`** |
| 2 | **`.`** | แทนอักขระใดก็ได้ 1 ตัว ยกเว้นขึ้นบรรทัดใหม่ | **`c.t`** | **`cat cut c9t`** | **`cat`**, **`cut`**, **`c9t`** |
| 3 | **`^`** | จับคู่ที่จุดเริ่มต้นของสตริง | **`^Hello`** | **`Hello world`** | **`Hello`** |
| 4 | **`$`** | จับคู่ที่จุดสิ้นสุดของสตริง | **`end$`** | **`the end`** | **`end`** |
| 5 | **`*`** | ซ้ำ 0 ครั้งหรือมากกว่า | **`ab*c`** | **`ac abc abbc`** | **`ac`**, **`abc`**, **`abbc`** |
| 6 | **`+`** | ซ้ำ 1 ครั้งหรือมากกว่า | **`ab+c`** | **`ac abc abbc`** | **`abc`**, **`abbc`** |
| 7 | **`?`** | มีหรือไม่มีก็ได้ 1 ครั้ง | **`colou?r`** | **`color colour`** | **`color`**, **`colour`** |
| 8 | **`{}`** | กำหนดจำนวนครั้งที่ซ้ำ | **`\d{3}`** | **`ID 123 and 4567`** | **`123`**, **`456`** |
| 9 | **`()`** | จัดกลุ่ม pattern | **`(ha)+`** | **`ha haha hahaha`** | **`ha`**, **`haha`**, **`hahaha`** |
| 10 | **`\`** | escape อักขระพิเศษ หรือใช้กับ special sequence | **`\.`** | **`version 3.10`** | **`.`** |
| 11 | **`|`** | เลือกอย่างใดอย่างหนึ่ง (or) | **`cat\|dog`** | **`cat and dog`** | **`cat`**, **`dog`** |


### 1. `[]` - วงเล็บเหลี่ยม

วงเล็บเหลี่ยมใช้กำหนดชุดอักขระที่ต้องการจับคู่

| นิพจน์ | สตริง | ตรงกันหรือไม่ | 
|:----| :--- |:--- |
|             |  **`a`** | ตรงกัน 1 ครั้ง | 
|             |  **`ac`** | ตรงกัน 2 ครั้ง |
| **`[abc]`** |  **`Hey Jude`** | ไม่ตรงกัน |
|             |  **`abc de ca`** | ตรงกัน 5 ครั้ง |

ในที่นี้ **`[abc]`** จะจับคู่ถ้าสตริงที่กำลังตรวจสอบมี **`a`**, **`b`** หรือ **`c`** ตัวใดตัวหนึ่ง

ตัวอย่างเพิ่มเติม:

| Pattern | ข้อความ | ผลที่จับคู่ได้ |
|:---|:---|:---|
| **`[aeiou]`** | **`Python regex is fun`** | **`o`**, **`e`**, **`e`**, **`i`**, **`u`** |
| **`[123]`** | **`Room 204, floor 3`** | **`2`**, **`3`** |
| **`[Pp]ython`** | **`Python and python`** | **`Python`**, **`python`** |

คุณยังสามารถกำหนดช่วงของอักขระได้โดยใช้ **`-`** ภายในวงเล็บเหลี่ยม

* **`[a-c]`** หมายถึง a หรือ b หรือ c
* **`[a-z]`** หมายถึงตัวอักษรใดก็ได้ตั้งแต่ a ถึง z
* **`[A-Z]`** หมายถึงอักขระใดก็ได้ตั้งแต่ A ถึง Z
* **`[0-3]`** หมายถึง 0 หรือ 1 หรือ 2 หรือ 3
* **`[0-9]`** หมายถึงตัวเลขใดก็ได้ตั้งแต่ 0 ถึง 9
* **`[A-Za-z0-9]`** หมายถึงอักขระเดี่ยวใดก็ได้ ได้แก่ a ถึง z, A ถึง Z หรือ 0 ถึง 9

ตัวอย่างการใช้ช่วงอักขระ:

| Pattern | ข้อความ | ผลที่จับคู่ได้ |
|:---|:---|:---|
| **`[a-c]`** | **`cat dog`** | **`c`**, **`a`** |
| **`[a-z]+`** | **`Python 3 Regex`** | **`ython`**, **`egex`** |
| **`[A-Z]+`** | **`Python REGEX 101`** | **`P`**, **`REGEX`** |
| **`[0-9]+`** | **`A12 B7 C300`** | **`12`**, **`7`**, **`300`** |
| **`[A-Za-z0-9]+`** | **`user_01@test`** | **`user`**, **`01`**, **`test`** |

คุณสามารถกลับความหมายของชุดอักขระได้โดยใช้สัญลักษณ์ caret **`^`** ที่จุดเริ่มต้นของวงเล็บเหลี่ยม

* **`[^abc]`** หมายถึงอักขระใดก็ได้ ยกเว้น **`a`**, **`b`** หรือ **`c`**
* **`[^0-9]`** หมายถึงอักขระที่ไม่ใช่ตัวเลข

ตัวอย่างการกลับความหมายของชุดอักขระ:

| Pattern | ข้อความ | ผลที่จับคู่ได้ |
|:---|:---|:---|
| **`[^abc]+`** | **`abcXYZ123`** | **`XYZ123`** |
| **`[^0-9]+`** | **`Order 123 shipped`** | **`Order `**, **` shipped`** |
| **`[^A-Za-z ]+`** | **`Price is $25.50!`** | **`$25.50!`** |

### 3. `^` - เครื่องหมายแคเร็ต

สัญลักษณ์ caret **`^`** ใช้ตรวจสอบว่าสตริง **ขึ้นต้นด้วย** อักขระบางตัวหรือไม่

  * **`r'^substring'`** เช่น **`r'^love'`** หมายถึงประโยคที่ขึ้นต้นด้วยคำว่า love
  * **`r'[^abc]'`** หมายถึงไม่ใช่ a, ไม่ใช่ b, ไม่ใช่ c เมื่อใช้ภายในวงเล็บเหลี่ยม

| นิพจน์ | สตริง | ตรงกันหรือไม่ | 
|:----| :--- |:--- |
|          |  **`a`** | ตรงกัน 1 ครั้ง | 
| **`^a`** |  **`abc`** | ตรงกัน 1 ครั้ง |
|          |  **`bac`** | ไม่ตรงกัน |
|----------|------------|------------------------------------------------------------|
| **`^a`** |  **`abc`** | ตรงกัน 1 ครั้ง |
|          |  **`acd`** | ตรงกัน 1 ครั้ง เพราะขึ้นต้นด้วย **`a`** |

ตัวอย่างเพิ่มเติมเมื่อใช้ `^` เพื่อจับจุดเริ่มต้นของสตริง:

| Pattern | ข้อความ | ผลที่จับคู่ได้ |
|:---|:---|:---|
| **`^Hello`** | **`Hello Python`** | **`Hello`** |
| **`^Python`** | **`I love Python`** | ไม่ตรงกัน |
| **`^\d+`** | **`2026 report`** | **`2026`** |
| **`^[A-Z]`** | **`Bangkok`** | **`B`** |

ตัวอย่างเมื่อใช้ `^` ภายใน `[]` เพื่อกลับความหมาย:

| Pattern | ข้อความ | ผลที่จับคู่ได้ |
|:---|:---|:---|
| **`[^0-9]+`** | **`abc123`** | **`abc`** |
| **`[^aeiou]`** | **`cat`** | **`c`**, **`t`** |

### 8. `{}` - ปีกกา

เราสามารถกำหนดความยาวของ substring ที่ต้องการค้นหาในข้อความได้โดยใช้ปีกกา **`{}`**

พิจารณาโค้ดนี้: **`{n,m}`** หมายถึงทำซ้ำอย่างน้อย **`n`** ครั้ง และมากที่สุด **`m`** ครั้ง ของรูปแบบที่อยู่ทางซ้าย

* **`{3}`**: ตรง 3 อักขระพอดี
* **`{3,}`**: อย่างน้อย 3 อักขระ
* **`{3,8}`**: 3 ถึง 8 อักขระ

| นิพจน์ | สตริง | ตรงกันหรือไม่ | 
|:----| :--- |:--- |
|              |  **`abc dat`** | ไม่ตรงกัน | 
|              |  **`abc daat`** | ตรงกัน 1 ครั้ง (ที่ **`daat`**) |
| **`a{2,3}`** |  **`aabc daaat`** | ตรงกัน 2 ครั้ง (ที่ **`aabc`** และ **`daaat`**) |
|              |  **`aabc daaaat`** | ตรงกัน 2 ครั้ง (ที่ **`aabc`** และ **`daaaat`**) |

ตัวอย่างการกำหนดจำนวนครั้ง:

| Pattern | ความหมาย | ข้อความ | ผลที่จับคู่ได้ |
|:---|:---|:---|:---|
| **`\d{3}`** | ตัวเลข 3 หลักพอดี | **`ID 1234 and 567`** | **`123`**, **`567`** |
| **`a{2,}`** | ตัว **`a`** อย่างน้อย 2 ตัว | **`a aa aaa`** | **`aa`**, **`aaa`** |
| **`a{2,4}`** | ตัว **`a`** ตั้งแต่ 2 ถึง 4 ตัว | **`a aa aaaa aaaaa`** | **`aa`**, **`aaaa`**, **`aaaa`** |
| **`[A-Z]{2}\d{3}`** | ตัวพิมพ์ใหญ่ 2 ตัว ตามด้วยตัวเลข 3 หลัก | **`AB123 X9 CD456`** | **`AB123`**, **`CD456`** |

ลองดูอีกหนึ่งตัวอย่าง RegEx **`[0-9]{2, 4}`** นี้จับคู่ตัวเลขอย่างน้อย 2 หลัก แต่ไม่เกิน 4 หลัก

| นิพจน์ | สตริง | ตรงกันหรือไม่ | 
|:----| :--- |:--- |
|                  |  **`ab123csde`** | ตรงกัน 1 ครั้ง (จับคู่ที่ **`ab123csde`**) | 
| **`[0-9]{2,4}`** |  **`12 and 345673`** | ตรงกัน 3 ครั้ง (**`12`**, **`3456`**, **`73`**) |
|                  |  **`1 and 2`** | ไม่ตรงกัน |


### 9. `|` - Alternation

เส้นตั้ง **`|`** ใช้สำหรับ alternation หรือตัวดำเนินการ **`or`**

| นิพจน์ | สตริง | ตรงกันหรือไม่ | 
|:----| :--- |:--- |
|           |  **`cde`** | ไม่ตรงกัน | 
| **`a\|b`** |  **`ade`** | ตรงกัน 1 ครั้ง (จับคู่ที่ **`a`**) |
|           |  **`acdbea`** | ตรงกัน 3 ครั้ง (**`a`**, **`b`**, **`a`**) |

ในที่นี้ **`a\|b`** จับคู่สตริงใดก็ได้ที่มี **`a`** หรือ **`b`**

ตัวอย่างเพิ่มเติม:

| Pattern | ข้อความ | ผลที่จับคู่ได้ |
|:---|:---|:---|
| **`cat\|dog`** | **`cat and dog`** | **`cat`**, **`dog`** |
| **`yes\|no`** | **`yes, maybe, no`** | **`yes`**, **`no`** |
| **`Python\|JavaScript`** | **`Python and JavaScript`** | **`Python`**, **`JavaScript`** |
| **`\d{3}\|[A-Z]{2}`** | **`AB 123 xyz`** | **`AB`**, **`123`** |


### 10. `()` - กลุ่ม

วงเล็บ **`()`** ใช้จัดกลุ่ม sub-patterns เช่น **`(a\|b\|c)xz`** จับคู่สตริงที่ตรงกับ **`a`**, **`b`** หรือ **`c`** แล้วตามด้วย **`xz`**

| นิพจน์ | สตริง | ตรงกันหรือไม่ | 
|:----| :--- |:--- |
|                 |  **`ab xz`** | ไม่ตรงกัน | 
| **`(a\|b\|c)xz`** |  **`abxz`** | ตรงกัน 1 ครั้ง (จับคู่ที่ **`bxz`**) |
|                 |  **`axz cabxz`** | ตรงกัน 2 ครั้ง (**`axz`**, **`bxz`**) |

ตัวอย่างเพิ่มเติม:

| Pattern | ข้อความ | ผลที่จับคู่ได้ |
|:---|:---|:---|
| **`(ha)+`** | **`ha haha hahaha`** | **`ha`**, **`haha`**, **`hahaha`** |
| **`(cat\|dog)s?`** | **`cat cats dog dogs`** | **`cat`**, **`cats`**, **`dog`**, **`dogs`** |
| **`(\d{3})-(\d{4})`** | **`Tel 123-4567`** | **`123-4567`** และ subgroup คือ **`123`**, **`4567`** |
| **`(ab){2,3}`** | **`ab abab ababab`** | **`abab`**, **`ababab`** |


### 11. `\` - Backslash

Backslash **`\`** ใช้ escape อักขระหลายชนิด รวมถึง metacharacters ทั้งหมด ตัวอย่างเช่น

**`\$a`** จะจับคู่ถ้าสตริงมี **`$`** ตามด้วย **`a`** ในที่นี้ **`$`** จะไม่ถูก RegEx engine ตีความแบบพิเศษ

ถ้าคุณไม่แน่ใจว่าอักขระตัวใดมีความหมายพิเศษหรือไม่ คุณสามารถใส่ **`\`** ไว้ข้างหน้าได้ วิธีนี้ช่วยให้แน่ใจว่าอักขระนั้นจะไม่ถูกตีความแบบพิเศษ

ตัวอย่างเพิ่มเติม:

| Pattern | ความหมาย | ข้อความ | ผลที่จับคู่ได้ |
|:---|:---|:---|:---|
| **`\.`** | จุดจริง ไม่ใช่ “อักขระใดก็ได้” | **`version 3.10`** | **`.`** |
| **`\$\d+`** | เครื่องหมายดอลลาร์ตามด้วยตัวเลข | **`Price is $250`** | **`$250`** |
| **`\(test\)`** | วงเล็บจริงรอบคำว่า test | **`run (test)`** | **`(test)`** |
| **`\\`** | backslash จริง 1 ตัว | **`C:\Users\Name`** | **`\`** |
| **`\d+`** | special sequence สำหรับตัวเลข 1 หลักขึ้นไป | **`ID 456`** | **`456`** |

### 1. `re.findall()`

เมธอด **`re.findall()`** คืนค่า list ของสตริงที่มีการจับคู่ทั้งหมด

````python
# ตัวอย่าง 1: re.findall()

# โปรแกรมสำหรับดึงตัวเลขจากสตริง

import re

string = 'hello 12 hi 89. Howdy 34'
pattern = '\d+'

result = re.findall(pattern, string) 
print(result)

# ผลลัพธ์: ['12', '89', '34']
````

ถ้าไม่พบรูปแบบ **`re.findall()`** จะคืนค่า list ว่าง

````python
# ตัวอย่าง 2: re.findall()

txt = '''Python is the most beautiful language that a human being has ever created.
I recommend python for a first programming language'''

# คืนค่าเป็น list
matches = re.findall('language', txt, re.I)
print(matches)  # ['language', 'language']
````

อย่างที่เห็น คำว่า **`language`** ถูกพบสองครั้งในสตริง มาฝึกเพิ่มเติมกัน
ตอนนี้เราจะค้นหาทั้งคำว่า **`Python`** และ **`python`** ในสตริง:

````python
# ตัวอย่าง 3: re.findall()

txt = '''Python is the most beautiful language that a human being has ever created.
I recommend python for a first programming language'''

# คืนค่าเป็น list
matches = re.findall('python', txt, re.I)
print(matches)  # ['Python', 'python']
````

เนื่องจากเราใช้ **`re.I`** จึงรวมทั้งตัวพิมพ์เล็กและตัวพิมพ์ใหญ่ หากไม่มี flag **`re.I`** เราจะต้องเขียน pattern ต่างออกไป ลองตรวจสอบดู:

````python
# ตัวอย่าง 4: re.findall()

txt = '''Python is the most beautiful language that a human being has ever created.
I recommend python for a first programming language'''

matches = re.findall('Python|python', txt)
print(matches)  # ['Python', 'python']

matches = re.findall('[Pp]ython', txt)
print(matches)  # ['Python', 'python']
````

### 2. `re.split()`

เมธอด **`re.split`** แยกสตริงตรงตำแหน่งที่มีการจับคู่ และคืนค่า list ของสตริงหลังการแยก

````python
# ตัวอย่าง 1: re.split()

txt = '''I am teacher and  I love teaching.
There is nothing as rewarding as educating and empowering people.
I found teaching more interesting than any other jobs.
Does this motivate you to be a teacher?'''
print(re.split('\n', txt)) # แยกโดยใช้ \n - สัญลักษณ์จบบรรทัด
````

````python
# ตัวอย่าง 2: re.split()

import re

string = 'Twelve:12 Eighty nine:89.'
pattern = '\d+'

result = re.split(pattern, string) 
print(result)

# ผลลัพธ์: ['Twelve:', ' Eighty nine:', '.']
````

ถ้าไม่พบรูปแบบ **`re.split`** จะคืนค่า list ที่มีสตริงเดิม

คุณสามารถส่งอาร์กิวเมนต์ **`maxsplit`** ให้เมธอด **`re.split`** ได้ ซึ่งเป็นจำนวนครั้งสูงสุดที่จะทำการแยก

````python
# ตัวอย่าง 3: re.split()

import re

string = 'Twelve:12 Eighty nine:89 Nine:9.'
pattern = '\d+'

# maxsplit = 1
# แยกเฉพาะตำแหน่งแรกที่พบ
result = re.split(pattern, string, 1) 
print(result)

# ผลลัพธ์: ['Twelve:', ' Eighty nine:89 Nine:9.']
````

อีกอย่างหนึ่ง ค่าเริ่มต้นของ **`maxsplit`** คือ 0 หมายถึงแยกทุกตำแหน่งที่เป็นไปได้

### 3. `re.sub()`

**ไวยากรณ์:**

```python
re.sub(pattern, replace, string)
```

เมธอดนี้คืนค่าสตริงที่การจับคู่ถูกแทนที่ด้วยเนื้อหาของตัวแปร **`replace`**

````python
# ตัวอย่าง 1: re.sub()

# โปรแกรมสำหรับลบ whitespace ทั้งหมด
import re

# สตริงหลายบรรทัด
string = 'abc 12\
de 23 \n f45 6'

# จับคู่ whitespace ทุกตัว
pattern = '\s+'

# สตริงว่าง
replace = ''

new_string = re.sub(pattern, replace, string) 
print(new_string)

# ผลลัพธ์: abc12de23f456
````

ถ้าไม่พบรูปแบบ **`re.sub()`** จะคืนค่าสตริงเดิม

คุณสามารถส่ง **`count`** เป็นพารามิเตอร์ตัวที่สี่ให้เมธอด **`re.sub()`** ได้ หากละไว้ ค่าจะเป็น 0 ซึ่งหมายถึงแทนที่ทุกครั้งที่พบ

````python
# ตัวอย่าง 2: re.sub()

import re

# สตริงหลายบรรทัด
string = 'abc 12\
de 23 \n f45 6'

# จับคู่ whitespace ทุกตัว
pattern = '\s+'
replace = ''

new_string = re.sub(r'\s+', replace, string, 1) 
print(new_string)

# ผลลัพธ์:
# abc12de 23
# f45 6
````

````python
# ตัวอย่าง 3: re.sub()

txt = '''Python is the most beautiful language that a human being has ever created.
I recommend python for a first programming language'''

match_replaced = re.sub('Python|python', 'JavaScript', txt, re.I)
print(match_replaced)  # JavaScript is the most beautiful language that a human being has ever created.
# หรือ
match_replaced = re.sub('[Pp]ython', 'JavaScript', txt, re.I)
print(match_replaced)  # JavaScript is the most beautiful language that a human being has ever created.
````

เพิ่มอีกหนึ่งตัวอย่าง สตริงต่อไปนี้อ่านยากมากหากไม่ลบสัญลักษณ์ **`%`** การแทนที่ **`%`** ด้วยสตริงว่างจะช่วยทำความสะอาดข้อความ

````python
# ตัวอย่าง 4: re.sub()

txt = '''%I a%m te%%a%%che%r% a%n%d %% I l%o%ve te%ach%ing. 
T%he%re i%s n%o%th%ing as r%ewarding a%s e%duc%at%i%ng a%n%d e%m%p%ow%er%ing p%e%o%ple.
I fo%und te%a%ching m%ore i%n%t%er%%es%ting t%h%an any other %jobs. 
D%o%es thi%s m%ot%iv%a%te %y%o%u to b%e a t%e%a%cher?'''

matches = re.sub('%', '', txt)
print(matches)
````

### 4. `re.subn()`

**`re.subn()`** คล้ายกับ **`re.sub()`** แต่จะคืนค่า tuple 2 รายการ ได้แก่สตริงใหม่และจำนวนครั้งที่แทนที่

````python
# ตัวอย่าง 1: re.subn()

# โปรแกรมสำหรับลบ whitespace ทั้งหมด
import re

# สตริงหลายบรรทัด
string = 'abc 12\
de 23 \n f45 6'

# จับคู่ whitespace ทุกตัว
pattern = '\s+'

# สตริงว่าง
replace = ''

new_string = re.subn(pattern, replace, string) 
print(new_string)

# ผลลัพธ์: ('abc12de23f456', 4)
````

### 5. `re.search()`

เมธอด **`re.search()`** รับอาร์กิวเมนต์สองตัวคือ pattern และ string เมธอดนี้ค้นหาตำแหน่งแรกที่ pattern ของ RegEx จับคู่กับสตริงได้

ถ้าค้นหาสำเร็จ **`re.search()`** จะคืนค่า match object หากไม่สำเร็จจะคืนค่า **`None`**

```python
match = re.search(pattern, str)
```

````python
# ตัวอย่าง 1: re.search()

import re

string = "Python is fun"

# ตรวจสอบว่า 'Python' อยู่ที่จุดเริ่มต้นหรือไม่
match = re.search('\APython', string)

if match:
    print("พบ pattern ภายในสตริง")
else:
    print("ไม่พบ pattern")  

# ผลลัพธ์: พบ pattern ภายในสตริง
````

ในที่นี้ **`match`** มี match object อยู่

````python
# ตัวอย่าง 2: re.search()

import re

txt = '''Python is the most beautiful language that a human being has ever created.
I recommend python for a first programming language'''

# คืนค่า object ที่มี span และ match
match = re.search('first', txt, re.I)
print(match)  # <re.Match object; span=(100, 105), match='first'>
# เราสามารถรับตำแหน่งเริ่มต้นและสิ้นสุดของ match เป็น tuple ได้ด้วย span
span = match.span()
print(span)     # (100, 105)
# หาตำแหน่งเริ่มต้นและสิ้นสุดจาก span
start, end = span
print(start, end)  # 100 105
substring = txt[start:end]
print(substring)       # first
````

## Match object

คุณสามารถดูเมธอดและแอตทริบิวต์ของ match object ได้โดยใช้ฟังก์ชัน **[dir()](https://github.com/milaan9/04_Python_Functions/blob/main/002_Python_Functions_Built_in/016_Python_dir().ipynb)**

เมธอดและแอตทริบิวต์ที่ใช้บ่อยของ match object ได้แก่:

````python
import re

txt = 'I love to teach python and javaScript'
# คืนค่า object ที่มี span และ match
match = re.match('I love to teach', txt, re.I)
print(match)  # <re.Match object; span=(0, 15), match='I love to teach'>
# เราสามารถรับตำแหน่งเริ่มต้นและสิ้นสุดของ match เป็น tuple ได้ด้วย span
span = match.span()
print(span)     # (0, 15)
# หาตำแหน่งเริ่มต้นและสิ้นสุดจาก span
start, end = span
print(start, end)  # 0, 15
substring = txt[start:end]
print(substring)       # I love to teach
````

จากตัวอย่างด้านบน pattern ที่เรากำลังค้นหา (หรือ substring ที่เรากำลังค้นหา) คือ **`I love to teach`** ฟังก์ชัน match จะคืนค่า object **เฉพาะเมื่อ** ข้อความขึ้นต้นด้วย pattern เท่านั้น

````python
import re

txt = 'I love to teach python and javaScript'
match = re.match('I like to teach', txt, re.I)
print(match)  # None
````

สตริงไม่ได้ขึ้นต้นด้วย **`I like to teach`** ดังนั้นจึงไม่มีการจับคู่ และเมธอด match คืนค่า **`None`**

### 1. `match.group()`

เมธอด **`group()`** คืนค่าส่วนของสตริงที่เกิดการจับคู่

````python
# ตัวอย่าง 6: Match object

import re

string = '39801 356, 2102 1111'

# ตัวเลขสามหลัก ตามด้วยช่องว่าง ตามด้วยตัวเลขสองหลัก
pattern = '(\d{3}) (\d{2})'

# ตัวแปร match มี Match object อยู่
match = re.search(pattern, string) 

if match:
    print(match.group())
else:
    print("ไม่พบ pattern")

# ผลลัพธ์: 801 35
````

re.findall() = หา “ทุกตัวที่เจอ” แล้วคืนค่าเป็น list

ในที่นี้ ตัวแปร match มี match object อยู่

## การใช้ prefix `r` ก่อน RegEx

เมื่อใช้ prefix **`r`** หรือ **`R`** ก่อน regular expression จะหมายถึง raw string ตัวอย่างเช่น **`'\n'`** คือบรรทัดใหม่ ในขณะที่ **`r'\n'`** หมายถึงอักขระสองตัว: backslash **`\`** ตามด้วย **`n`**

Backslash **`\`** ใช้ escape อักขระหลายชนิด รวมถึง metacharacters ทั้งหมด อย่างไรก็ตาม การใช้ prefix **`r`** ทำให้ **`\`** ถูกปฏิบัติเหมือนอักขระปกติ

````python
# ตัวอย่าง 7: Raw string using r prefix

import re

string = '\n and \r are escape sequences.'

result = re.findall(r'[\n\r]', string) 
print(result)

# ผลลัพธ์: ['\n', '\r']
````

## ตัวอย่าง RegEx กับอักขระพิเศษ (Metacharacters)

มาดูตัวอย่างเพื่ออธิบาย อักขระพิเศษ ด้วยเมธอด RegEx กัน:

### วงเล็บเหลี่ยม `[]`

ลองใช้วงเล็บเหลี่ยมเพื่อรวมทั้งตัวพิมพ์เล็กและตัวพิมพ์ใหญ่

````python
# ตัวอย่าง 1:

regex_pattern = r'[Aa]pple' # วงเล็บเหลี่ยมนี้หมายถึง A หรือ a
txt = 'Apple and banana are fruits. An old cliche says an apple a day a doctor way has been replaced by a banana a day keeps the doctor far far away.'
matches = re.findall(regex_pattern, txt)
print(matches)  # ['Apple', 'apple']
````

ถ้าเราต้องการค้นหาทั้ง apple และ banana เราจะเขียน pattern ดังนี้:

````python
# ตัวอย่าง 2:

regex_pattern = r'[Aa]pple|[Bb]anana' # วงเล็บเหลี่ยมนี้หมายถึง A หรือ a
txt = 'Apple and banana are fruits. An old cliche says an apple a day a doctor way has been replaced by a banana a day keeps the doctor far far away.'
matches = re.findall(regex_pattern, txt)
print(matches)  # ['Apple', 'banana', 'apple', 'banana']
````

ด้วยการใช้วงเล็บเหลี่ยม **`[]`** และตัวดำเนินการ or **`|`** เราสามารถดึง Apple, apple และ Banana ออกมาได้

### อักขระ escape `\`

````python
# ตัวอย่าง 1:

regex_pattern = r'\d'  # d เป็นอักขระพิเศษที่หมายถึงตัวเลข
txt = "Hawking born on 8 January 1942 and died on 14 March 2018 Einstein's birth anniversary (Pi-Day) and both died at 76"
matches = re.findall(regex_pattern, txt)
print(matches)  # ['8', '1', '9', '4', '2', '1', '4', '2', '0', '1', '8', '7', '6'] - นี่ยังไม่ใช่สิ่งที่เราต้องการ
````

### หนึ่งครั้งหรือมากกว่า `+`

````python
# ตัวอย่าง 1:

regex_pattern = r'\d+'  # d เป็นอักขระพิเศษที่หมายถึงตัวเลข และ + หมายถึงหนึ่งครั้งหรือมากกว่า
txt = "Hawking born on 8 January 1942 and died on 14 March 2018 Einstein's birth anniversary (Pi-Day) and both died at 76"
matches = re.findall(regex_pattern, txt)
print(matches)  # ['8', '1942', '14', '2018', '76'] - แบบนี้ดีกว่า
````

### จุด `.`

````python
# ตัวอย่าง 1:

regex_pattern = r'[a].'  # วงเล็บเหลี่ยมนี้หมายถึง a และ . หมายถึงอักขระใดก็ได้ยกเว้นบรรทัดใหม่
txt = '''Apple and Banana are fruits'''
matches = re.findall(regex_pattern, txt)
print(matches)  # ['an', 'an', 'an', 'a ', 'ar']
````

````python
# ตัวอย่าง 2: [] with +

regex_pattern = r'[a].+'  # . หมายถึงอักขระใดก็ได้ และ + หมายถึงหนึ่งครั้งหรือมากกว่า 
matches = re.findall(regex_pattern, txt)
print(matches)  # ['and Banana are fruits']
````

### ศูนย์ครั้งหรือมากกว่า `*`

````python
# ตัวอย่าง 1:

regex_pattern = r'[a].*'  # . หมายถึงอักขระใดก็ได้ และ * หมายถึงศูนย์ครั้งหรือมากกว่า 
txt = '''Apple and Banana are fruits'''
matches = re.findall(regex_pattern, txt)
print(matches)  # ['and Banana are fruits']
````

### ศูนย์ครั้งหรือหนึ่งครั้ง `?`

ศูนย์ครั้งหรือหนึ่งครั้ง หมายถึง pattern อาจไม่เกิดขึ้นเลย หรือเกิดขึ้นหนึ่งครั้งก็ได้

````python
# ตัวอย่าง 1:

txt = '''I am not sure if there is a convention how to write the word e-mail.
Some people write it as email others may write it as Email or E-mail.'''
regex_pattern = r'[Ee]-?mail'  # ? ในที่นี้หมายถึง '-' จะมีหรือไม่มีก็ได้
matches = re.findall(regex_pattern, txt)
print(matches)  # ['e-mail', 'email', 'Email', 'E-mail']
````

### Quantifier `{}`

เราสามารถกำหนดความยาวของ substring ที่ต้องการค้นหาในข้อความได้โดยใช้ปีกกา **`{}`** ลองสมมติว่าเราสนใจ substring ที่มีความยาว 4 อักขระ:

````python
# ตัวอย่าง 1:

txt = "Hawking born on 8 January 1942 and died on 14 March 2018 Einstein's birth anniversary (Pi-Day) and both died at 76"
regex_pattern = r'\d{4}'  # ตรงสี่ครั้งพอดี
matches = re.findall(regex_pattern, txt)
print(matches)  # ['1942', '2018']
````

````python
# ตัวอย่าง 2:

txt = "Hawking born on 8 January 1942 and died on 14 March 2018 Einstein's birth anniversary (Pi-Day) and both died at 76"
regex_pattern = r'\d{1,4}'   # 1 ถึง 4
matches = re.findall(regex_pattern, txt)
print(matches)  # ['8', '1942', '76', '14', '2018', '76']
````

### เครื่องหมายแคเร็ต `^`

````python
# ตัวอย่าง 1: ขึ้นต้นด้วย

txt = "Hawking born on 8 January 1942 and died on 14 March 2018 Einstein's birth anniversary (Pi-Day) and both died at 76"
regex_pattern = r'^Hawking'  # ^ หมายถึงขึ้นต้นด้วย
matches = re.findall(regex_pattern, txt)
print(matches)  # ['Hawking']
````

````python
# ตัวอย่าง 2: Negation

txt = "Hawking born on 8 January 1942 and died on 14 March 2018 Einstein's birth anniversary (Pi-Day) and both died at 76"
regex_pattern = r'[^A-Za-z ]+'  # ^ ในชุดอักขระหมายถึงการปฏิเสธ ไม่ใช่ A ถึง Z, ไม่ใช่ a ถึง z และไม่ใช่ช่องว่าง
matches = re.findall(regex_pattern, txt)
print(matches)  # ['8', '1942', '14', '2018', "'", '(', '-', ')', '76']
````

### เครื่องหมายดอลลาร์ `$`

เครื่องหมาย **`$`** ใช้จับคู่ที่จุดสิ้นสุดของสตริง

````python
# ตัวอย่าง 1: ลงท้ายด้วย

txt = 'I like Python'
regex_pattern = r'Python$'  # $ หมายถึงต้องอยู่ท้ายสตริง
matches = re.findall(regex_pattern, txt)
print(matches)  # ['Python']
````

````python
# ตัวอย่าง 2: ไม่ตรงกัน เพราะคำว่า Python ไม่ได้อยู่ท้ายสตริง

txt = 'Python is fun'
regex_pattern = r'Python$'
matches = re.findall(regex_pattern, txt)
print(matches)  # []
````

### Alternation `|`

เครื่องหมาย **`|`** หมายถึง “หรือ” ใช้เลือกจับ pattern อย่างใดอย่างหนึ่ง

````python
# ตัวอย่าง 1: cat หรือ dog

txt = 'I have a cat and a dog'
regex_pattern = r'cat|dog'
matches = re.findall(regex_pattern, txt)
print(matches)  # ['cat', 'dog']
````

````python
# ตัวอย่าง 2: Python หรือ JavaScript

txt = 'Python and JavaScript are programming languages'
regex_pattern = r'Python|JavaScript'
matches = re.findall(regex_pattern, txt)
print(matches)  # ['Python', 'JavaScript']
````

### กลุ่ม `()`

วงเล็บ **`()`** ใช้จัดกลุ่ม pattern เพื่อให้ metacharacter ตัวอื่นทำงานกับทั้งกลุ่ม หรือเพื่อแยก subgroup ที่จับคู่ได้

````python
# ตัวอย่าง 1: ทำซ้ำทั้งกลุ่ม ha

txt = 'ha haha hahaha 34454 7654'
# regex_pattern = r'(?:\d)+'  # (?:...) คือกลุ่มที่ไม่ต้องการเก็บ subgroup
regex_pattern = r'(?:ha)+'  # (?:...) คือกลุ่มที่ไม่ต้องการเก็บ subgroup
matches = re.findall(regex_pattern, txt)
print(matches)  # ['ha', 'haha', 'hahaha']
````

````python
# ตัวอย่าง 2: แยก subgroup จากเบอร์โทรศัพท์

txt = 'Tel: 123-4567'
regex_pattern = r'(\d{3})-(\d{4})'
match = re.search(regex_pattern, txt)

if match:
    print(match.group())   # 123-4567
    print(match.group(1))  # 123
    print(match.group(2))  # 4567
````

### Backslash `\`

Backslash **`\`** ใช้ escape อักขระพิเศษ เพื่อบอกว่าเราต้องการจับตัวอักษรนั้นจริง ๆ ไม่ใช่ใช้ความหมายพิเศษของ RegEx

````python
# ตัวอย่าง 1: จับจุดจริง ๆ ด้วย \.

txt = 'version 3.10'
regex_pattern = r'\.'  # \. หมายถึงจุดจริง ไม่ใช่อักขระใดก็ได้
matches = re.findall(regex_pattern, txt)
print(matches)  # ['.']
````

````python
# ตัวอย่าง 2: จับเครื่องหมาย $ จริง ๆ

txt = 'Price is $250'
regex_pattern = r'\$\d+'  # \$ หมายถึงเครื่องหมาย $ จริง และ \d+ หมายถึงตัวเลข 1 ตัวขึ้นไป
matches = re.findall(regex_pattern, txt)
print(matches)  # ['$250']
````

## โจทย์ฝึก MetaCharacters

โจทย์ชุดนี้ออกแบบให้ฝึก metacharacter ทีละตัวแบบไม่ยากมาก ให้เขียน RegEx pattern และใช้ `re.findall()`, `re.search()` หรือ `re.match()` ตามความเหมาะสม โดยพยายามใช้ metacharacter ของหัวข้อนั้นเป็นแกนหลัก

คำใบ้: ข้อที่ค้นหาจากหลายบรรทัด เช่น `^` และ `$` สามารถใช้ `re.MULTILINE` ช่วยได้

### 1. `[]` - ตู้ล็อกเกอร์สีรุ้ง

มีรหัสล็อกเกอร์ของนักเรียนหลายคน ให้ดึงเฉพาะรหัสที่ขึ้นต้นด้วยสี `R`, `G` หรือ `B` แล้วตามด้วยตัวเลข 2 หลัก

```python
text = 'R12 A09 G07 B88 Y42 R5 B03'
# ผลลัพธ์ที่คาดหวัง: ['R12', 'G07', 'B88', 'B03']
```

### 2. `.` - รหัสเมนูลับ

ร้านกาแฟมีรหัสเมนู 3 ตัวอักษรที่ขึ้นต้นด้วย `C` และลงท้ายด้วย `T` โดยตัวกลางเป็นอะไรก็ได้ 1 ตัว ให้ค้นหารหัสที่ตรงเงื่อนไข

```python
text = 'CAT C7T C-T CT COFFEE CUT'
# ผลลัพธ์ที่คาดหวัง: ['CAT', 'C7T', 'C-T', 'CUT']
```

### 3. `^` - งานที่ต้องเริ่มวันนี้

จากรายการงานหลายบรรทัด ให้ดึงบรรทัดที่ขึ้นต้นด้วยคำว่า `TODAY:` เท่านั้น

```python
text = '''TODAY: submit report
LATER: clean dataset
TODAY: review regex
DONE: install package'''
# ผลลัพธ์ที่คาดหวัง: ['TODAY: submit report', 'TODAY: review regex']
```

### 4. `$` - ไฟล์ส่งอาจารย์

จากรายชื่อไฟล์หลายบรรทัด ให้ดึงเฉพาะไฟล์ที่ลงท้ายด้วย `.pdf`

```python
text = '''summary.pdf
notes.txt
regex_homework.pdf
image.pdf.bak
slide.PDF'''
# ผลลัพธ์ที่คาดหวัง: ['summary.pdf', 'regex_homework.pdf']
```

### 5. `*` - เสียงปรบมือในแชต

ในแชตมีคนพิมพ์ `clap`, `claap`, `claaap` หรือบางคนพิมพ์ `clp` โดยลืมตัว `a` ให้ดึงคำเหล่านี้ทั้งหมด

```python
text = 'clp clap claap claaap cloooap clip'
# ผลลัพธ์ที่คาดหวัง: ['clp', 'clap', 'claap', 'claaap']
```

### 6. `+` - เสียงหัวเราะที่ต้องมีเสียงจริง

ให้ดึงคำหัวเราะที่เป็น `ha` แล้วมีตัว `!` อย่างน้อย 1 ตัว เช่น `ha!`, `ha!!`, `ha!!!`

```python
text = 'ha ha! ha!! ha!!! haha!'
# ผลลัพธ์ที่คาดหวัง: ['ha!', 'ha!!', 'ha!!!']
```

### 7. `?` - สะกดได้สองแบบ

````python
import re

paragraph = '''I love teaching. If you do not love teaching what else can you love. I love Python if you do not love something which can give you all the capabilities to develop an application what else can you love.'''

words = re.findall(r'\b\w+\b', paragraph)
word_counts = {}
print(words)

for word in words:
    word_counts[word] = word_counts.get(word, 0) + 1

most_frequent = sorted(
    [(count, word) for word, count in word_counts.items()],
    reverse=True
)

print(most_frequent)
print('คำที่พบบ่อยที่สุด:', most_frequent[0])
````

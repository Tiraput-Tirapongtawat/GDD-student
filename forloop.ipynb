## ข้อ 1: นับถอยหลังปล่อยจรวด

ให้ใช้ `for` พิมพ์ตัวเลขถอยหลังจาก `start` ไปถึง 1 แล้วพิมพ์ `Launch!`

ตัวอย่าง: ถ้า `start = 5` ต้องได้

```text
5
4
3
2
1
Launch!
```

````python
start = 5
for i in range(start,0, -1): #start stop step
    print(i)
print("Launch!")
# for i in range (start,stop,step)
````

## ข้อ 2: ป้ายชื่อเข้าค่าย

มีรายชื่อใน list `names` ให้พิมพ์ป้ายชื่อในรูปแบบ `Hello, ชื่อ!` ทุกคน

เช่น `Hello, Mina!`

## ข้อ 3: รวมพลังเหรียญในเกม

มี list `coins` เก็บจำนวนเหรียญที่เก็บได้แต่ละด่าน ให้ใช้ `for` รวมเหรียญทั้งหมด แล้วพิมพ์ `Total coins: ...`

````python
coins = [10, 25, 5, 40, 20]
total = 0
for i in coins:
    print(i)
    # total = total + i
    total += i
print(total)
````

## ข้อ 4: ตรวจคะแนนที่ผ่านเกณฑ์

มี list `scores` ให้พิมพ์เฉพาะคะแนนที่มากกว่าหรือเท่ากับ 50

จากนั้นพิมพ์จำนวนคนที่ผ่าน เช่น `Passed: 3 people`

````python
scores = [72, 48, 91, 33, 50, 64]
````

````python
people = 0
for i in scores:
    if i > 50:
        people += 1
print("Pass : " + str(people))
    
````

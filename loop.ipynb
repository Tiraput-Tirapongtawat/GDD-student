# While Loop

while condition:
    method ......

```python
# run จนกว่า count จะมากกว่า 5 (count > 5) condition : ควรเป็น true เสมอ จนกว่า count จะมากกว่า 5 (count > 5 -> False)
count = 0

while  count < 5:
    count = count + 1
    print("Start")
```

```python
count = 3
while count < 5 :
    print(count)
    count = count + 1
else:
    print("Finished task , count : " , count)
```

```python
password = "1234"
retry = 0
    
while retry < 3:
    your_password  = input("Insert your password.")
    if password == your_password:
        print( "Complete Login")
        break
    else:
        retry = retry + 1
        print("Incorrect password please try again")
else:
    print("Login Failed please try again in 3 hours")
```

```python
a = 1

if a == 2:
    print("test1")

if a == 1:
    print("test2")
else:
    print("test3")
```

```python
count = 0
while count < 5:
    if count != 3:
        print(count)
        count = count + 1
    else:
        count = count + 1
        
```

```python
count = 0
while count < 5:
    count = count + 1
    print(count)
    if count == 3:
        continue
```

```python
count = 0
while count < 5:
    if count == 3:
        count = count + 1
        continue
    print(count)
    count = count + 1
```

# For loop

for ตัววนซ้ำ in ข้อมูล:
    วิธีการ.....

```python
# for loop with list
list_example = [1,3,4,5,6,77,4]

for i in list_example:
    print(i)
```

```python
# for loop with string
string_example = "sf bsj shdfb"

for i in string_example:
    print(i)
```

```python
numbers = [1,3,4,4,5,3,23,4,234,23,42,34,100,345]

for number in numbers:
    print(number)
    if number ==  100:
        print("We will break when number = 100")
        break
    
```

```python
for i in range(0,101):
    if i%7 == 0:
        continue
    print(i)
```

```python
for i in range(3):
    print(i)
else:
    print("Task done")
```

```python
for i in range(30):
    print(i)
    if i == 2:
        print("break")
        break
else:
    print("Task done")
```

```python
nums = [1, 2, 3, 4, 5, 6,9,12,67,80,67,79,101]
# ให้ print ตัวเลขทุกตัวที่ไม่ใช่เลขคี่ 
# หยุด เมื่อเจอเลขที่หาร 7 ลงตัว
# กรณีที่ไม่มีเลขที่หาร 7 ลงตัวเลย ให้ print Not found

for i in nums:
    # print(i)
    if i%2 == 0 :
        print(i)
    elif i%7 == 0:
        # print(i)
        break
if i%7 != 0 :
    print("Not found")
        
```

```python
nums = [1, 2, 3, 4, 5, 6, 9, 12, 67, 80, 67, 79, 101]
# ให้ print ตัวเลขทุกตัวที่ไม่ใช่เลขคี่
# หยุด เมื่อเจอเลขที่หาร 7 ลงตัว
# กรณีที่ไม่มีเลขที่หาร 7 ลงตัวเลย ให้ print Not found

for i in nums:
    if i % 2 != 0:
        continue
    elif i % 7 == 0:
        break
    print(i)
else:
    print("Not found")
```
